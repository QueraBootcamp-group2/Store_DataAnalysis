## Statistics ##


```python
# Import libraries 
import plotly.express as px
import pandas as pd
from plotly.subplots import make_subplots
from scipy.stats import normaltest, ttest_ind, shapiro
pd.set_option('display.max_columns', None)
```

```python
# Data sets
df = pd.read_csv('order_detail.csv')
with_discount = df[df['Discount'] != 0]
without_discount = df[df['Discount'] == 0]
with_discount['Type'] = 'With Discount'
without_discount['Type'] = 'Without Discount'
```

```python
prods_with = with_discount.groupby('Product ID', as_index=False)['Quantity'].sum()
prods_with
```

```python
prods_without = without_discount.groupby('Product ID', as_index=False)['Quantity'].sum()
prods_without
```

```python
# Functions
def Histogram(df1, df2):
    fig = px.histogram(x= df1, y= df2, marginal= 'box', nbins= 50)
    return fig

def Boxplot(df1, df2):
    fig = px.box(x = df1, y = df2, points= 'all', notched= True)
    return fig

def CheckNormality(x):
    statistic, pvalue = normaltest(x)
    if pvalue > 0.05:
        print(f"{x} is normally distributed.")
    else:
        print(f"{x} is not normally distributed.")
    print(f"The Normality test results:\n the statistic result is {statistic}. The greater the value, the farther away it is from being normal. \n the pvalue result is {pvalue:.4f}.")
    return

def CheckNormality2(x):
    statistic, pvalue = shapiro(x)
    if pvalue > 0.05:
        print(f"{x} is normally distributed.")
    else:
        print(f"{x} is not normally distributed.")
    print(f"The Normality test results:\n the statistic result is {statistic}. The closer it is to 1, the more normal the data is. \n the pvalue result is {pvalue:.4f}.")
    return

def Parametric(x, y):
    res = ttest_ind(x, y, equal_var= True)
    print(f"The result of the non parametric test is: ")
    print(f"The statistic is : {res.statistic:.4f}. This value shows how many standard errors the means are apart from each other.")
    print(f"The pvalue of the test is: {res.pvalue:.10f}")
    print(f"The degrees of freedom is: {res.df:.4f}")
    if res.pvalue < 0.05:
        print(f"There is a statistical significance between Quantity and their Discounts.")
    else:
        print(f"There is no statistically significant connection between Quantity and their Discounts. Hence we were unable to reject the null hypothesis.")
    return res 

def NonParametric(x, y):
    res = ttest_ind(x, y)
    print(f"The result of the non parametric test is: ")
    print(f"The statistic is : {res.statistic:.4f}. This value shows how many standard errors the means are apart from each other.")
    print(f"The pvalue of the test is: {res.pvalue:.10f}")
    print(f"The degrees of freedom is: {res.df:.4f}")
    if res.pvalue < 0.05:
        print(f"There is a statistical significance between Quantity and their Discounts.")
    else:
        print(f"There is no statistically significant connection between Quantity and their Discounts. Hence we were unable to reject the null hypothesis.")
    return res
```

```python
# Visualisation
fig = make_subplots(rows=1, cols=2, subplot_titles=['With Discount', 'Without Discount'])
for trace in Histogram(prods_with['Product ID'], prods_with['Quantity']).data:
    fig.add_trace(trace, row=1, col=1)

for trace in Histogram(prods_without['Product ID'], prods_without['Quantity']).data:
    fig.add_trace(trace, row=1, col=2)

fig.update_layout(height=600, width= 1400, title_text="Histogram of Sales by their Quantity", legend_title= 'Sales')
fig.show()
```

```python
prods_with['Type'] = 'With Discount'
prods_without['Type'] = 'Without Discount'
combined = pd.concat([prods_with, prods_without], ignore_index=True)
fig = px.histogram(combined, x= 'Quantity', color= 'Type', opacity= 0.6, barmode= 'overlay', nbins= 50, title= 'Summed Quantity Distribution: With vs Without Discount')
fig.update_layout(xaxis_title= 'Total Quantity per Product', yaxis_title= 'Count of Products',legend_title= 'Group')
fig.show()
```
